```python
import pandas as pd

# Load the data
data = pd.read_csv("headphone.csv")

# Display the first few rows of the data
data.head()
```

```python
import pandas as pd
import nltk
from nltk.corpus import words

# Assuming you've already loaded your data into a dataframe named mouse_data

# Create a set of English words
english_words = set(words.words())

def is_english(text):
    # Ensure the text is a string
    if not isinstance(text, str):
        return False
    # Tokenize the text
    tokens = nltk.word_tokenize(text)
    # If there are no tokens, return False
    if not tokens:
        return False
    # Count how many tokens are in the English words set
    english_count = sum(1 for token in tokens if token.lower() in english_words)
    # If more than 70% of words are English, consider it an English text
    return english_count / len(tokens) > 0.7

# Filter the dataframe
data = data[data['Review Content'].apply(is_english)]
```

```python
data = data.dropna()
```

```python
data
```

```python
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification
from scipy.special import softmax
```

```python
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
import torch
# Ensure model is in eval mode
model.eval()

def get_sentiment_scores(review_text):
    # Tokenize input and get output from the model
    encoded_input = tokenizer.encode(review_text, return_tensors='pt', truncation=True, max_length=512)
    with torch.no_grad():
        output = model(encoded_input)

    # Convert logits to probabilities
    probs = softmax(output.logits.numpy(), axis=1).flatten()

    # Create a dictionary with sentiment scores
    sentiment_scores = {
        'neg': probs[0],
        'neu': probs[1],
        'pos': probs[2]
    }
    return sentiment_scores

# Apply sentiment analysis to filtered reviews
data['Sentiment Scores'] = data['Review Content'].apply(get_sentiment_scores)

# Display results
```

```python
print(data[['Review Content', 'Sentiment Scores']])
```

```python
# Extract the 'pos' value from 'Sentiment Scores' and create a new column 'Positive Score'
data['Positive Score'] = data['Sentiment Scores'].apply(lambda x: x['pos'])

# Sort by the positive sentiment score in descending order and take the top 5
top_5_positive_reviews = data.sort_values(by='Positive Score', ascending=False).head(5)

# Display the top 5 positive reviews with their link and image
top_5_positive_reviews[['Review Content', 'Sentiment Scores', 'Positive Score', 'Link']]
```

```python
# Check if there is an "Unnamed" column that can be renamed to "ID"
unnamed_columns = [col for col in data.columns if "Unnamed" in col]

# If "Unnamed" column exists, rename it to "ID". Otherwise, create an "ID" column.
if unnamed_columns:
    
    data.rename(columns={unnamed_columns[0]: 'ID'}, inplace=True)
else:
    data.insert(0, 'ID', range(1, len(data) + 1))

# Show the first few rows to confirm changes
data.head()
```

```python
data['Review Content'].fillna('', inplace=True)
```

```python
data['Review Content'] = data['Review Content'].str.strip()
```

```python
data.to_csv("headphone_final.csv")
```
